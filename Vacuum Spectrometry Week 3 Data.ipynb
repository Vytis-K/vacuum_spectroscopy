```python
#data for Xe-129 isotope

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# data
xdata = [128.2,128.3,128.4,128.5,128.6,128.7,128.8]
ydata = [0.0000000188,0.0000000235,0.0000000321,0.0000000418,0.0000000399,0.0000000321,0.0000000244]

x = np.asarray(xdata)
y = np.asarray(ydata)

# error bars
error = y * 0.02

# define the Gaussian function
def gaussian(x, a, x0, sigma):
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2))

# provide initial guess for parameters
max_y = np.max(y)
max_x = x[np.argmax(y)]
initial_guess = [max_y, max_x, 0.05]

# graph data and Gaussian

popt , pcov = curve_fit(gaussian, x, y, initial_guess)
  
plt.plot(x, y, 'b.', label='data')
plt.plot(x, gaussian (x , *popt), 'r-', label='fit')
plt.errorbar(x, y, yerr=error, fmt='o')
plt.legend()

plt.xlabel('AMU')
plt.ylabel('Pressure (Torr)')
plt.title('Xe-129 Peak')

plt.show()

plt.savefig('xe129.png')

# calculate maximum value and uncertainty
max_val = gaussian(popt[1], *popt)
max_err = np.sqrt(pcov[1,1])

print('Maximum value:', max_val)
print('Uncertainty in maximum value:', max_err)
```
